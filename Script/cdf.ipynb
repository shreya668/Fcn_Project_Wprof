```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import scipy.stats as ss 

def plotcdf(fname, label1, key):
    fin = open(fname, 'r')
    import matplotlib.pyplot as plt

    ar = []
    for line in fin.readlines():
        lsplit = line.split(',')
        if str(lsplit[1]) == key and float(lsplit[2]) > 0: 
            ar.append(float(lsplit[2]))


    sorted_data = np.sort(ar)   
    yvals=np.arange(len(sorted_data))/float(len(sorted_data))

    plt.plot(sorted_data,yvals, label=label1)
    plt.ylabel('cdf')
    plt.xlabel('Time(ms)')
    #plt.title('Desktop - Cdf of Time Download')
    #plt.savefig('desk_cdf_time_download_without.png')

    

plotcdf('new/new_1.csv', 'Cold Cache', 'time_download')
plotcdf('new/new_3.csv', 'Warm Cache', 'time_download')
plotcdf('new/new_6.csv', 'Hot Cache', 'time_download')


plt.title('Desktop - Cdf of Time Download')
plt.legend(loc='lower right')
plt.savefig('new.png')
plt.show()
```
